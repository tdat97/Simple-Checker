```python
import numpy as np
import json

def json2label(path): # json 경로
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)
    labels = [shape["label"] for shape in data["shapes"]]
    points = [np.array(shape["points"]).astype(np.int32)
             for shape in data["shapes"]] # (n, 2?, 2)
    return dict(zip(labels, points))

label_dic = json2label("./temp/test.json")
label_dic
```

```python
import numpy as np
import cv2

def get_poly_box_wh(poly_box): # (4, 2)
    lt, rt, rb, lb = poly_box
    w = int((np.linalg.norm(lt - rt) + np.linalg.norm(lb - rb)) // 2)
    h = int((np.linalg.norm(lt - lb) + np.linalg.norm(rt - rb)) // 2)
    return w, h

def crop_obj_in_bg(bg_img, polys):
    obj_imgs = []
    for poly in polys:
        poly = poly.astype(np.float32)
        w, h = get_poly_box_wh(poly)
        pos = np.float32([[0,0], [w,0], [w,h], [0,h]])
        M = cv2.getPerspectiveTransform(poly, pos)
        obj_img = cv2.warpPerspective(bg_img, M, (w, h))
        obj_imgs.append(obj_img)
    return obj_imgs

img = cv2.imread("./temp/test.jpg")
crop_img = crop_obj_in_bg(img, label_dic["test"][None])
```

```python
import cv2

def img_clahe(img):
    b, g, r = cv2.split(img)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    b = clahe.apply(b)
    g = clahe.apply(g)
    r = clahe.apply(r)
    new_img = cv2.merge((b,g,r))
    return new_img

new_img = img_clahe(crop_img[0])
```

```python
import os
import json
import numpy as np

JSON_DIR_PATH = "./source/json"

def load_jsons(dir_path):
    file_names = os.listdir(dir_path)
    json_paths = list(map(lambda x:os.path.join(dir_path, x), file_names))
    names = list(map(lambda x:x.split('.')[0], file_names))
    # open json
    polys_list = []
    for path in json_paths:
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
        points = np.float32([shape["points"] for shape in data["shapes"]])
        polys_list.append(points)
    return dict(zip(names, polys_list))

aa = load_jsons(JSON_DIR_PATH)['1234'][0]

aa
```
